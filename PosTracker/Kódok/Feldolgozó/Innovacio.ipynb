```python
plt.plot(df['t'][0:200]-df['t'][0:200].shift(periods=1))
plt.xlabel("Mintavétel száma [-]")
plt.ylabel("Mintavételhez szükséges idő [s]")
plt.title("Mintavételezési idő")
```

```python
minta_v = np.nan_to_num(np.array(df['t']-df['t'].shift(periods=1)))
minta_v = minta_v[1:]
print(np.mean(minta_v))
print(np.min(minta_v))
print(np.max(minta_v))
print(np.std(minta_v))
from scipy.stats import gaussian_kde
data = minta_v
density = gaussian_kde(data)
xs = np.linspace(0,0.05,200)
density.covariance_factor = lambda : .25
density._compute_covariance()
plt.plot(xs,density(xs))
plt.show()
```
